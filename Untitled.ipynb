```python
import pandas as pd
import numpy as np

# 读取 CSV 文件
file_path = 'rowdata.csv'  # 替换为你的文件路径
df = pd.read_csv(file_path)

# 选择前三列
input_matrix = df.iloc[:, :3].values

def compute_lengths(points):
    lengths = [0]
    steps = []
    for i in range(1, len(points)):
        length = np.linalg.norm(points[i] - points[i-1])
        lengths.append(lengths[-1] + length)
        steps.append(length)
    lengths = np.round(lengths, 3)
    steps = np.round(steps, 3)
    return lengths, steps

lengths, steps = compute_lengths(input_matrix)
print(lengths)

# 将结果转换为 DataFrame
result_df = pd.DataFrame({
    'Lengths': lengths
    
})

# 写入新的 CSV 文件
output_file_path = 'output.csv'  # 替换为你希望保存的文件路径
result_df.to_csv(output_file_path, index=False)

print(output_file_path)
print(steps)
```
