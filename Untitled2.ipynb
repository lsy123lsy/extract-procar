```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#  CSV 
data = pd.read_csv('final.csv')

#  400 
group_size = 400
num_groups = len(data) // group_size + (1 if len(data) % group_size != 0 else 0)

# 
plt.figure(figsize=(12, 8))

for i in range(num_groups):
    # 
    group_data = data.iloc[i*group_size:(i+1)*group_size]

    #  x, y, size 
    x = group_data['path']
    y = group_data['afterfermi']
    sizes = group_data['tot'] * 40  # 

    # 
    plt.scatter(x, y, s=sizes, facecolors='none', edgecolors='blue', alpha=0.5, label=f'Group {i+1}')

    # 
    plt.plot(x, y, label=f'Line {i+1}', linewidth=1, color='black')  

#  y 
plt.ylim(-15, 8)
plt.xlim(0, 2.49)
#  x 
x_ticks_labels = {'K': 0, 'G': 0.918, 'L': 1.779, 'W': 2.129, 'X': 2.486}
plt.xticks(list(x_ticks_labels.values()), list(x_ticks_labels.keys()))
# 
#plt.legend()
plt.xlabel('Path')
plt.ylabel('After Fermi')
plt.title('Bubble and Line Plot-tot')
# 
plt.savefig('bubble_line_plot_tot.png')
plt.show()
```
